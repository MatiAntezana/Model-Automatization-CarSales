## *Analizamos la frecuancia de los diferenes features del **DataFrame**.*

### **Marcas de Automoviles**

```python
marca_counts = df['Marca'].value_counts()
plt.figure(figsize=(20, 10)) 
num_colors = len(marca_counts)
colors = plt.cm.hsv(np.linspace(0, 1, num_colors))
bars = plt.bar(marca_counts.index, marca_counts.values, color=colors)
plt.xticks(rotation=90)
plt.title('Frecuencia de las marcas de los vehículos', fontsize=30)
plt.xlabel('Marca', fontsize=20)
plt.ylabel('Publicaciones', fontsize=20)
legend_labels = [f"{marca}: {count}" for marca, count in zip(marca_counts.index, marca_counts.values)]
plt.legend(bars, legend_labels, title="Marcas y Cantidad de Autos", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## ***Cambiar el formato de los precios de los automoviles a dolares.***

Establecemos el la **conversión de dolares** a utilizar.

```python
dolar_blue = 1040 # U$S1 = $1040
```

Aplicamos la **conversión a los precios** de los automoviles y **eliminamos** la columna del tipo de moneda.

```python
# Aplicamos la conversión de precios a dólares
df['Precio'] = df.apply(lambda row: round(row['Precio'] / dolar_blue) if row['Moneda'] == '$' else round(row['Precio']), axis=1)

# Renombramos la columna 'Precio' a 'Precio (U$S)'
df = df.rename(columns={'Precio': 'Precio (U$S)'})

# Eliminamos la columna 'Moneda'
df = df.drop(columns=['Moneda'])

# Mostramos los cambios aplicado
df
```

## ***Eliminación de columnas innecesarias***

Eliminamos la feature ***Tipo de carrocería*** ya que todos los datos del set son **SUV**. 

```python
df = df = df.drop(columns=['Tipo de carrocería'])
df
```

Eliminamos la feature ***Con cámara de retroceso*** ya que son **muchas las muestras que no nos aportan este dato** (ademas de que no se concidera muy necesario a la hora de establecer el precio del automovil)

```python
df = df = df.drop(columns=['Con cámara de retroceso'])
df
```
